## 5. Data Cleaning

```python
print("Duplicate Rows Before:", df.duplicated().sum())
```

```python
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df["date_added"] = pd.to_datetime(
    df["date_added"].str.strip(),
    format="mixed"
)
```

```python
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month_name()
```

```python
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
```

```python
df.to_csv("../data/cleaned_netflix_titles.csv", index=False)
```

## 6. Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df, x="type")

plt.title("Movies vs TV Shows")

plt.savefig("../images/type_distribution.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="year_added",
    order=df["year_added"].value_counts().index
)

plt.xticks(rotation=90)

plt.title("Content Added to Netflix by Year")

plt.savefig("../images/content_added_by_year.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
top_country = df["country"].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_country.values,
    y=top_country.index
)

plt.title("Top 10 Countries Producing Netflix Content")

plt.savefig("../images/top_countries.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    y="rating",
    order=df["rating"].value_counts().index
)

plt.title("Content Ratings")

plt.savefig("../images/content_ratings.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df["release_year"],
    bins=30,
    kde=True
)

plt.title("Release Year Distribution")

plt.savefig("../images/release_year_distribution.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
top_directors = df[df["director"] != "Unknown"]["director"].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_directors.values,
    y=top_directors.index
)

plt.title("Top 10 Directors")

plt.savefig("../images/top_directors.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
genres = (
    df["listed_in"]
    .str.split(", ")
    .explode()
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10,5))

sns.barplot(
    x=genres.values,
    y=genres.index
)

plt.title("Top 10 Genres")

plt.savefig("../images/top_genres.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(
    data=df,
    x="year_added",
    hue="type"
)

plt.xticks(rotation=90)

plt.title("Movies and TV Shows Added Each Year")

plt.savefig("../images/movies_vs_tvshows_year.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="month_added",
    order=month_order
)

plt.xticks(rotation=45)

plt.title("Monthly Content Added")

plt.savefig("../images/monthly_content_added.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

```python
df["type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.ylabel("")

plt.title("Movies vs TV Shows Percentage")

plt.savefig("../images/type_percentage.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

### Key Business Insights

- Movies make up the majority of Netflix's content library.
